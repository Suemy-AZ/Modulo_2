Eligiremos solo las variables que tengan una correlación moderada que se considera arriba de 0.3 y dividimos los datos en 70% entrenamiento y 30% para pruebas

```python
from sklearn.model_selection import train_test_split

#x=df[["INDUS","NOX","AGE","DIS","RAD","TAX","B","LSTAT","MEDV"]]
x=df[["DIS","MEDV","B","ZN","PTRATIO","LSTAT","AGE","CRIM","CRIM","NOX","INDUS","RAD"]]
y=df[["TAX"]]

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.7)
```

Elegimos el modelo de: .

```python
from sklearn import ensemble
from sklearn.tree import DecisionTreeClassifier

myTree = DecisionTreeClassifier()
myTree.fit(x_train, y_train)


myForest= ensemble.RandomForestClassifier()
myForest.fit(x_train,y_train)
```

```python
from sklearn.metrics import r2_score, confusion_matrix

print(myForest.score(x_test,y_test))
print(myTree.score(x_test,y_test))

# Calculamos el coeficiente de determinación
CoD = r2_score(y_test, myForest.predict(x_test))
CoD2 = r2_score(y_test, myTree.predict(x_test))
print(CoD)
print(CoD2)

```
